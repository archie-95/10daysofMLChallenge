# Cleaning of Data

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='tab20b_r')
```

### This shows that Age and cabin has Missing values


### As we cant autofill cabin values, but we can auto fill the Age Values 

```python
#this shows the box plot of age based on Pclass
plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass',y='Age',data=train)
```

### now we find out the median of Age of people , grouped by Pclass

```python
#Selecting rows having Pclass=1 in the DataSet
rslt_df_pclass_1 = train.loc[(train['Pclass'] == 1)]
rslt_df_pclass_1
```

```python
#Selecting rows having Pclass=2 in the DataSet
rslt_df_pclass_2 = train.loc[(train['Pclass'] == 2)]
rslt_df_pclass_2
```

```python
#Selecting rows having Pclass=3 in the DataSet
rslt_df_pclass_3 = train.loc[(train['Pclass'] == 3)]
rslt_df_pclass_3
```

```python
#to get the number of rows
x,y=train.shape
x
```

### Now we will replace the NaN values to the median of that particular group

```python
for i in range(x):
    if np.isnan(train.iloc[i][5]):
        if train.iloc[i][2]==1:
            train.iat[i,5]=rslt_df_pclass_1['Age'].median()
        elif train.iloc[i][2]==2:
            train.iat[i,5]=rslt_df_pclass_2['Age'].median()
        elif train.iloc[i][2]==3:
            train.iat[i,5]=rslt_df_pclass_3['Age'].median()
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='magma')
plt.title("The NaN values is Age is replaced by the Median of age")
plt.show()
```

### As Cabin no is not a useful information , we will drop the column

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='tab20b_r')
plt.title("The Cabin column is dropped")
plt.show()
```
